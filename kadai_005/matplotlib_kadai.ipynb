```python
# step1 matplotlib.pyplotnumpy
import matplotlib.pyplot as plt
import numpy as np

# step2 6
fig, axes = plt.subplots(nrows=2, ncols=3,tight_layout = True )

# step3 sin
#
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

fig, ax = plt.subplots()

axes[0,1].plot(x, y, linewidth=2.0)

axes[0,1].set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))

# 
x = np.random .normal(15, 5, 2000)
axes[0,2].hist(x)

# 
hertz = [300, 200, 200, 500, 600, 2000, 1600, 1800, 1850, 1700, 1500, 1500, 1500, 1400, 5000, 6000, 2000, 3000, 3000]
# 
axes[1,0].boxplot(hertz)

#step4 
plt.show()
```
